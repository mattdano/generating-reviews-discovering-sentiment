```python
from IPython.core.display import HTML
import matplotlib.cm as cm

def colour_by_score(message,scores):
    scaled_score = (scores+1)/2

    
    output = ''
    for x,w in enumerate(message):
        output+="<span style='background-color:rgba"+str(tuple(map(lambda x:int(x*255),cm.Spectral(scaled_score[x]))))+";'>"+w+'</span>'
    return output
```

```python
sequence_features = model.sequence_features(text[0],2388)
formatted_string = colour_by_score(text[0],sequence_features)
HTML(formatted_string)
```
